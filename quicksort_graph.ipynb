```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("quicksort_data.txt", sep=" ", header=None, names=["n", "time"])
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["n"], df["time"], marker="o", linestyle="-", linewidth=2, markersize=6)

plt.xlabel("Input Size (n)")
plt.ylabel("Time (seconds)")
plt.title("Quick Sort Performance: Time vs Input Size")
plt.grid(True)
plt.show()
```
